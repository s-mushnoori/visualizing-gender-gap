```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']
fig = plt.figure(figsize=(16, 20))

#Plots for STEM majors
for sp in range(0,18,3):
    major_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[major_index]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[major_index]], c=cb_orange, linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(True)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(True)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(stem_cats[major_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 8, 'Women')
    elif sp == 15:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 35, 'Women')
        ax.tick_params(labelbottom="on")

#Plots for Liberal Arts majors
for sp in range(1,15,3):
    major_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[major_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[major_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(True)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(True)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(lib_arts_cats[major_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if sp == 1:
        ax.text(2002, 80, 'Women')
        ax.text(2005, 18, 'Men')
    elif sp == 13:
        ax.tick_params(labelbottom="on")

#Plots for Other majors
for sp in range(2,18,3):
    major_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[major_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[major_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(True)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(True)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[major_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if sp == 2:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif sp == 17:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 35, 'Women')
        ax.tick_params(labelbottom="on")
        
plt.savefig('gender_degrees.png')
plt.show()
```
